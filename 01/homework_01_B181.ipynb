```python
import pandas as pd
import matplotlib.pyplot as plt
from beakerx import *

data = pd.read_csv('output_final.csv', sep =';')
bachelor = data[data['Thesis_type']=='BP']
master = data[data['Thesis_type']=='DP']
%matplotlib inline
TableDisplay(data)
```

## Number of defended theses per year for CTU/Faculties

### Bachelor

```python
groupby_bach = bachelor.groupby(bachelor['Year'])
bach_df = pd.DataFrame({'Year':groupby_bach.size().index, 'Number of thesis':groupby_bach.size().values})
TableDisplay(bach_df)
# 2019 - chyba v systému, správně je 2016
```

### Master

```python
groupby_master = master.groupby(master['Year'])
master_df = pd.DataFrame({'Year':groupby_master.size().index, 'Number of thesis':groupby_master.size().values})
TableDisplay(master_df)
```

## Find the departments/study programmes/supervisors/reviewers with highest numbers of thesis and come up with some nice plots and tables to depict their numbers.

### Department

```python
data['Department_name'].value_counts().head(10).plot.bar()
plt.ylabel('Number of thesis')
plt.show()
```

### Program

```python
data['Program'].value_counts().head(10).plot.bar()
plt.ylabel('Number of thesis')
plt.show()
```

### Supervisor

```python
data['Supervisor'].value_counts().head(10).plot.bar()
plt.ylabel('Number of thesis')
plt.show()
```

### Reviewer

```python
without = data[data['Reviewer']!='-']
without['Reviewer'].value_counts().head(10).plot.bar()
plt.ylabel('Number of thesis')
plt.show()
```

## Mean/median/minimum/maximum number of supervised theses per year for faculties.

```python
faculties = data['Faculty_name'].unique()
o = OutputContainer()
for faculty in faculties:
    o.addItem(HTML(f"<h1 style=\"color: red;\">{faculty}</h1>"))
    per_years = data[data['Faculty_name']==faculty].groupby(['Year'])
    o.addItem(f'   Mean= {per_years.size().mean()}')
    o.addItem(f'   Median= {per_years.size().median()}')
    o.addItem(f'   Min= {per_years.size().min()}')
    o.addItem(f'   Max= {per_years.size().max()}')
    
    o.addItem('\n')
o
```
